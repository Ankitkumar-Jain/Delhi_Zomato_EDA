# Filtering Indian Restaurant basis of Country Code with the req columns

```python
new_df=df[df['Country Code']==1][['Restaurant ID','Restaurant Name','City','Address','Locality','Average Cost for two','Has Table booking',
                  'Has Online delivery','Is delivering now','Aggregate rating','Rating color','Rating text','Votes']]
```

```python
new_df.columns
```

# Renaming Column

```python
new_df.rename(columns={'Restaurant ID':'res_id','Restaurant Name':'res_name','City':'city','Address':'address',
                      'Locality':'locality','Average Cost for two':'avg_cost_for_two','Has Table booking':'has_table_booking',
                      'Has Online delivery':'has_online_delivery','Is delivering now':'is_delivering_now','Aggregate rating':'rating',
                      'Rating text':'review','Votes':'votes'},inplace=True)
```

```python
new_df.isnull().sum()
```

```python
new_df.sample(5)
```

# Total number of restaurant in City

```python
new_df['city'].value_counts().sort_values(ascending=False)
```

# Observation:

New Delhi has highest number of restaurant while Panchkula and Mohali have lowest number of restaurant

#  EDA on New Delhi City 

Filtering the data on the basis of city i.e. New Delhi

```python
nd = new_df[new_df['city']=='New Delhi']
```

```python
nd.sample(5)
```

```python
sns.catplot(data=nd,x='review',kind='count')
plt.show()
```

# Observation

above plot shows that there are more number of restaurant got Average review followed by Not rated and very less number of restaurant are excellent.

```python
ratings_df = nd.groupby(['rating','review','Rating color'])[['rating','review','Rating color']].size().reset_index().rename(columns={0:'count'})
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data = ratings_df,x='rating',y='count',hue='review')
plt.show()
```

# Observation

above plot shows that restaurant got rating between 2.5 to 3.4 ---> Average review , 3.5 to 3.9----->Good, 4.0 to 4.5----> Very Good , 4.6 to 4.9----> Excellent
